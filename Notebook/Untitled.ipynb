```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['Outcome'], axis=1)
y=df['Outcome']
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.linear_model import LogisticRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
des=DecisionTreeClassifier()
des2=LogisticRegression()
des3=KNeighborsClassifier(n_neighbors=2)
des4=LogisticRegression(penalty=None, solver='newton-cg')
## 'penalty': None, 'solver': 'newton-cg'
```

```python
des.fit(x_train,y_train)
des2.fit(x_train,y_train)
des3.fit(x_train,y_train)
des4.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred=des.predict(x_test)
y_pred2=des2.predict(x_test)
y_pred3=des3.predict(x_test)
y_pred4=des4.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred2))
print(classification_report(y_test,y_pred3))
print(classification_report(y_test,y_pred4))
```
